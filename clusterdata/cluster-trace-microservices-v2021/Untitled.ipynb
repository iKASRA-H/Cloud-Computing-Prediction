```python
import matplotlib.pyplot as plt
import pandas as pd

i = 0
for node in nodeid_unique:
    i += 1
    if i % 1000 == 0:
        series = pd.Series(df['nodeid'])
        indices = series[series == node].index.tolist()
        
        df_node = df.loc[indices, ['timestamp', 'node_cpu_usage', 'node_memory_usage']].copy()
        
        #  timestamp      (hh:mm:ss)
        df_node["time"] = pd.to_datetime(df_node["timestamp"], unit="ms").dt.strftime("%H:%M:%S")
        
        print(i)

        # ---  CPU Usage ---
        plt.figure(figsize=(10, 4))  #    
        plt.plot(df_node["time"], df_node["node_cpu_usage"], linestyle="-", label="CPU Usage", color="blue")

        plt.xlabel("Time (hh:mm:ss)")
        plt.ylabel("CPU Usage")
        plt.title(f"CPU Usage for Node {node}")
        plt.xticks(rotation=45)
        plt.legend()
        plt.grid()
        plt.show()

        # ---  Memory Usage ---
        plt.figure(figsize=(10, 4))  #    
        plt.plot(df_node["time"], df_node["node_memory_usage"], linestyle="-", label="Memory Usage", color="red")

        plt.xlabel("Time (hh:mm:ss)")
        plt.ylabel("Memory Usage")
        plt.title(f"Memory Usage for Node {node}")
        plt.xticks(rotation=45)
        plt.legend()
        plt.grid()
        plt.show()
```
