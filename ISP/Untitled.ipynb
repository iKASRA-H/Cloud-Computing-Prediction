```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data with proper datetime parsing
df = pd.read_csv(
    'ec_data_set.csv',
    skiprows=1,  # Skip header row containing "Time"
    header=None,
    names=['datetime', 'traffic_bits'],
    parse_dates=['datetime']
)

# Convert datetime column explicitly
df['datetime'] = pd.to_datetime(df['datetime'], format='%m/%d/%Y %H:%M')

# Set datetime as index
df.set_index('datetime', inplace=True)

# Create plot
plt.figure(figsize=(12, 6))
df['traffic_bits'].plot(color='blue', linewidth=0.5)

# Format plot
plt.title('Internet Traffic Over Time (5-minute intervals)')
plt.xlabel('Date and Time')
plt.ylabel('Traffic Volume (bits)')
plt.grid(True, which='both', linestyle='--', alpha=0.7)
plt.gcf().autofmt_xdate()

plt.tight_layout()
plt.show()
```
